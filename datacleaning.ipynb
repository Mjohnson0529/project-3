```python
#import dependecies 
import pandas as pd 
from pathlib import Path

# Make a reference to the csv file path
csv_path = Path("Resources/ObesityDataSet_raw_and_data_sinthetic.csv")


# Import the file as a DataFrame
obesity_df = pd.read_csv(csv_path)
obesity_df.head()
```

```python
# Remove columns
obesity_df = obesity_df[["Age",'Gender','Height', 'Weight','FCVC', 'NCP','family_history_with_overweight','FAF','CAEC','MTRANS','NObeyesdad']]
obesity_df.head()
obesity_renamed_df = obesity_df.rename(columns={"FCVC":"Frequency of vegetable consumption",'NCP': 'Number of main meals','Family history of overweight':'Family history','FAF':'Physical activity frequency','CAEC':'Consumption of food between meals','MTRANS': 'Transportation method', 'NObeyesdad': 'Obesity level' })
```

```python
# Rename colums 
obesity_renamed_df = obesity_df.rename(columns={'Height':'Height(in)','Weight':'Weight(lbs)', "FCVC":"Frequency of vegetable consumption",'NCP': 'Number of main meals','family_history_with_overweight':'Family history of obesity','FAF':'Physical activity frequency','CAEC':'Consumption of food between meals','MTRANS': 'Transportation method', 'NObeyesdad': 'Obesity level' })
obesity_renamed_df
```

```python
# Covert colums 
obesity_renamed_df['Height(in)'] = ((obesity_renamed_df['Height(in)'] * 3.28) * 12 ).round(2)
obesity_renamed_df['Weight(lbs)'] = (obesity_renamed_df['Weight(lbs)'] * 2.204).round(2)
obesity_renamed_df

# Add BMI column
obesity_renamed_df['BMI'] = ((obesity_renamed_df['Weight(lbs)']) / ((obesity_renamed_df['Height(in)'])**2)*703)
obesity_renamed_df['BMI'] = obesity_renamed_df['BMI'].round(2)
obesity_renamed_df

# Adjust age column
obesity_renamed_df['Age'] = obesity_renamed_df['Age'].round(0)

obesity_renamed_df
```

```python
# Create a for loop to change values in the columns 
for index, row in obesity_renamed_df.iterrows():
    if row['Consumption of food between meals'] == 'no':
        obesity_renamed_df.at[index,'Consumption of food between meals'] = 0
    if row['Consumption of food between meals'] == 'Sometimes':
        obesity_renamed_df.at[index,'Consumption of food between meals'] = 1
    if row['Consumption of food between meals'] == 'Frequently':
        obesity_renamed_df.at[index,'Consumption of food between meals'] = 2
    if row['Consumption of food between meals'] == 'Always':
        obesity_renamed_df.at[index,'Consumption of food between meals'] = 3

obesity_renamed_df
        
```
