```python
#import dependecies 
import pandas as pd 
from pathlib import Path

# import csv
cleaned_data = pd.read_csv("Resources/cleaned_obesity_dataout.csv")
```

```python
cleaned_data.head()
```

```python
df = cleaned_data.drop(columns=['Height(in)', 'Weight(lbs)', 'Family history of obesity', 'Transportation method'])
df
```

```python
Radar_df = df.groupby(['Obesity level','Gender']).mean().round(1)
Radar_df
```

```python
Radar_df.to_csv('Resources/obesity_radar_df.csv', index=True, header=True)
```

```python
df2 = cleaned_data.drop(columns=['Gender','Height(in)', 'Weight(lbs)', 'Family history of obesity', 'Transportation method'])
df2.head()
```
